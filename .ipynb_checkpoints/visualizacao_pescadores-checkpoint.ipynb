## 5. Pré-cálculo de estatísticas por estado e região

```python
por_estado    = df.groupby(['estado','regiao']).size().reset_index(name='total')
por_municipio = df.groupby(['estado','municipio']).size().reset_index(name='total')
por_regiao    = df.groupby('regiao').size().reset_index(name='total')

# Dicionário de stats por UF (injetado no JS do mapa)
stats_estado = {}
for _, row in por_estado.iterrows():
    uf        = row['estado']
    reg       = row['regiao']
    total_est = int(row['total'])
    total_reg = int(por_regiao[por_regiao['regiao']==reg]['total'].values[0])
    top_munis = (por_municipio[por_municipio['estado']==uf]
                 .sort_values('total', ascending=False).head(3))
    top_list  = [(r['municipio'], int(r['total'])) for _, r in top_munis.iterrows()]
    conc      = top_list[0][1] / total_est * 100 if top_list else 0
    stats_estado[uf] = {
        'total': total_est, 'regiao': reg, 'total_regiao': total_reg,
        'pct_brasil':  round(total_est / TOTAL_GERAL * 100, 1),
        'pct_regiao':  round(total_est / total_reg   * 100, 1),
        'top_municipios': top_list,
        'concentracao_capital': round(conc, 0),
    }

# Dicionário de stats por região
stats_regiao = {}
for _, row in por_regiao.iterrows():
    reg       = row['regiao']
    total_reg = int(row['total'])
    estados_r = por_estado[por_estado['regiao']==reg].sort_values('total', ascending=False)
    top_est   = [(r['estado'], int(r['total'])) for _, r in estados_r.iterrows()]
    top_mun   = (por_municipio[por_municipio['estado'].map(regioes_map)==reg]
                 .sort_values('total', ascending=False).head(3))
    top_mun_l = [(r['municipio'], int(r['total'])) for _, r in top_mun.iterrows()]
    stats_regiao[reg] = {
        'total': total_reg,
        'pct_brasil': round(total_reg / TOTAL_GERAL * 100, 1),
        'n_estados':  len(estados_r),
        'top_estados':    top_est,
        'top_municipios': top_mun_l,
    }

print('Stats calculadas:')
print(f'  {len(stats_estado)} estados | {len(stats_regiao)} regiões')
```

## 6. Download do GeoJSON dos estados brasileiros

```python
GEOJSON_URL = (
    'https://raw.githubusercontent.com/codeforamerica/click_that_hood'
    '/master/public/data/brazil-states.geojson'
)
estados_geojson = requests.get(GEOJSON_URL, timeout=30).json()

regioes_map_full = {
    'Acre':'Norte','Amazonas':'Norte','Amapá':'Norte','Pará':'Norte',
    'Rondônia':'Norte','Roraima':'Norte','Tocantins':'Norte',
    'Alagoas':'Nordeste','Bahia':'Nordeste','Ceará':'Nordeste','Maranhão':'Nordeste',
    'Paraíba':'Nordeste','Pernambuco':'Nordeste','Piauí':'Nordeste',
    'Rio Grande do Norte':'Nordeste','Sergipe':'Nordeste',
    'Distrito Federal':'Centro-Oeste','Goiás':'Centro-Oeste',
    'Mato Grosso do Sul':'Centro-Oeste','Mato Grosso':'Centro-Oeste',
    'Espírito Santo':'Sudeste','Minas Gerais':'Sudeste',
    'Rio de Janeiro':'Sudeste','São Paulo':'Sudeste',
    'Paraná':'Sul','Rio Grande do Sul':'Sul','Santa Catarina':'Sul',
}
nome_para_uf = {
    'Acre':'AC','Amazonas':'AM','Amapá':'AP','Pará':'PA','Rondônia':'RO',
    'Roraima':'RR','Tocantins':'TO','Alagoas':'AL','Bahia':'BA','Ceará':'CE',
    'Maranhão':'MA','Paraíba':'PB','Pernambuco':'PE','Piauí':'PI',
    'Rio Grande do Norte':'RN','Sergipe':'SE','Distrito Federal':'DF',
    'Goiás':'GO','Mato Grosso do Sul':'MS','Mato Grosso':'MT',
    'Espírito Santo':'ES','Minas Gerais':'MG','Rio de Janeiro':'RJ',
    'São Paulo':'SP','Paraná':'PR','Rio Grande do Sul':'RS','Santa Catarina':'SC',
}

total_por_uf  = dict(zip(por_estado['estado'],  por_estado['total']))
total_por_reg = dict(zip(por_regiao['regiao'],   por_regiao['total']))

# Enriquecer features com UF, região e contagens
for feat in estados_geojson['features']:
    nome = feat['properties'].get('name', '')
    uf   = nome_para_uf.get(nome, '')
    reg  = regioes_map_full.get(nome, 'Desconhecido')
    feat['properties']['uf']               = uf
    feat['properties']['regiao']           = reg
    feat['properties']['pescadores_estado'] = int(total_por_uf.get(uf, 0))
    feat['properties']['pescadores_regiao'] = int(total_por_reg.get(reg, 0))

print(f"GeoJSON carregado: {len(estados_geojson['features'])} estados")
```

## 7. Preparação dos dados de visualização por estado

Separa os pontos e dados de heatmap por estado — usados quando o usuário clica
num estado para filtrar o heatmap e os pontos apenas para aquela UF.

```python
# Pontos individuais amostrados por estado (máx 400 por UF)
pontos_por_estado = {}
for uf in df['estado'].unique():
    sub = df[df['estado']==uf].sample(min(400, len(df[df['estado']==uf])), random_state=42)
    pontos_por_estado[uf] = sub[['latitude','longitude','municipio']].values.tolist()

# Dados de heatmap por estado (todos os pontos)
heat_por_estado = {}
for uf in df['estado'].unique():
    sub = df[df['estado']==uf]
    heat_por_estado[uf] = sub[['latitude','longitude']].values.tolist()

# Heatmap do Brasil inteiro
heat_brasil = df[['latitude','longitude']].values.tolist()

print(f'Pontos por estado preparados: {len(pontos_por_estado)} UFs')
print(f'Heatmap Brasil: {len(heat_brasil)} pontos')
```

## 8. Geração do mapa HTML interativo

Esta célula monta o HTML completo com Leaflet.js embutido, injetando todos
os dados (GeoJSON, pontos, stats) como variáveis JavaScript.

### Funcionalidades implementadas:
- **Isolamento do Brasil**: fundo escuro `#1a1a2e`, tiles com opacidade 0, retângulo mundial sólido cobrindo o fundo
- **`maxBounds` + `maxBoundsViscosity: 1.0`**: impede arrastar o mapa para fora do Brasil
- **Clique no estado**: chama `selecionarEstado(uf)` que (1) reestiliza todos os polígonos deixando visível só o selecionado, (2) faz `fitBounds` com padding, (3) substitui o heatmap global pelo do estado, (4) atualiza o painel lateral
- **Botão "Voltar ao Brasil"**: chama `voltarBrasil()` que reverte tudo
- **Painel lateral**: 3 abas (Resumo, Ranking, Insights) com stats pré-calculadas em Python e inseridas como JSON no JS
- **Download TXT**: gera relatório de texto do estado/região selecionado

```python
COR_REGIAO = {
    'Norte':'#2E8B57', 'Nordeste':'#D4880E', 'Centro-Oeste':'#8B4513',
    'Sudeste':'#1565C0', 'Sul':'#6A0DAD', 'Desconhecido':'#888888',
}

# Serializar dados Python → JSON para injeção no JavaScript
estados_geojson_str = json.dumps(estados_geojson, ensure_ascii=False)
stats_estado_js     = json.dumps(stats_estado,    ensure_ascii=False)
stats_regiao_js     = json.dumps(stats_regiao,    ensure_ascii=False)
pontos_js           = json.dumps(pontos_por_estado, ensure_ascii=False)
heat_js             = json.dumps(heat_por_estado,   ensure_ascii=False)
heat_brasil_js      = json.dumps(heat_brasil,       ensure_ascii=False)

# Lê o template HTML e substitui os placeholders pelos dados reais
# O HTML usa Leaflet.js (CDN) + leaflet.heat para o heatmap
# Toda a lógica de interação está em JS puro dentro do HTML
import os
template_path = 'mapa_template.html'

# Se o template não existir, use o script build_map_v2.py para gerá-lo
# Alternativamente, execute build_map_v2.py diretamente:
if os.path.exists('build_map_v2.py'):
    exec(open('build_map_v2.py').read())
    print('Mapa gerado via build_map_v2.py')
else:
    print('AVISO: build_map_v2.py não encontrado.')
    print('Certifique-se de que build_map_v2.py está na mesma pasta que este notebook.')
    print('Ele é o script principal que gera mapa_pescadores_brasil.html')
```

## 10. Gráfico estático de resumo (para o relatório)

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

COR_REGIAO = {
    'Norte':'#2E8B57','Nordeste':'#D4880E','Centro-Oeste':'#8B4513',
    'Sudeste':'#1565C0','Sul':'#6A0DAD',
}

resumo = df.groupby('regiao').size().reset_index(name='total').sort_values('total', ascending=False)
top10  = (df.groupby(['estado','regiao']).size().reset_index(name='total')
            .sort_values('total', ascending=False).head(10))

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Distribuição de Armadores de Pesca no Brasil', fontsize=14, fontweight='bold')

# Por região
cores_reg = [COR_REGIAO[r] for r in resumo['regiao']]
axes[0].barh(resumo['regiao'], resumo['total'], color=cores_reg, edgecolor='white')
axes[0].set_xlabel('Número de pescadores registrados')
axes[0].set_title('Por Região')
for i, v in enumerate(resumo['total']):
    axes[0].text(v+20, i, str(v), va='center', fontsize=10)

# Top 10 estados
cores_top = [COR_REGIAO[r] for r in top10['regiao']]
axes[1].barh(top10['estado'], top10['total'], color=cores_top, edgecolor='white')
axes[1].set_xlabel('Número de pescadores registrados')
axes[1].set_title('Top 10 Estados')
for i, v in enumerate(top10['total']):
    axes[1].text(v+5, i, str(v), va='center', fontsize=9)

patches = [mpatches.Patch(color=c, label=r) for r, c in COR_REGIAO.items()]
axes[1].legend(handles=patches, title='Região', loc='lower right', fontsize=8)

plt.tight_layout()
plt.savefig('grafico_resumo_pescadores.png', dpi=150, bbox_inches='tight')
plt.show()
print('Gráfico salvo: grafico_resumo_pescadores.png')
```
